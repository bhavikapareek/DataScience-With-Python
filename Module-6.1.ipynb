# Data Manipulation

```python
dataset=dataset.rename(columns={'No. of Inmates benefitted by Elementary Education':'elementary', 
                               'No. of Inmates benefitted by Adult Education':'adult',
                               'No. of Inmates benefitted by Higher Education':'higher_ed',
                               'No. of Inmates benefitted by Computer Course':'computer'})
```

```python
dataset
```

```python
dataset['total_benefitted']=(dataset.elementary+dataset.adult+ dataset.higher_ed+ dataset.computer)
```

```python
dataset
```

```python
dataset["total_benefited"]=dataset.sum(axis=1)
```

```python
dataset
```

```python
import matplotlib.pyplot as plot
xlabels = dataset['STATE/UT'].values
plot.figure(figsize=(20, 3))
plot.xticks(np.arange(xlabels.shape[0]), xlabels, rotation = 'vertical', fontsize = 18)
plot.xticks
plot.bar(np.arange(dataset.values.shape[0]),dataset['total_benefited'],align = 'edge', color='green')
```
