```python
raw_data={'cust_id':[1,2,3,4,5], 
         'cust_name':['Alex', 'john', 'rose','jack', 'taylor'], 
         'state': ['MD','VA', 'FL','MA', 'NC'] }
```

```python
raw_data
```

```python
raw_data1={'c_id': [1,2,3,4,5], 
         'course':['python', 'tableau', 'r', 'excel', 'c'],
         'price': [233,200, 250,190, 250]}
```

```python
raw_data
```

```python
raw_data1
```

```python
import pandas as pd  ## importing pandas lib for creating a datafram
```

```python
customer_deatil=pd.DataFrame(raw_data) ## Converting dictionary to dataframes
```

```python
Sales_detail=pd.DataFrame(raw_data1)   ## Converting dictionary to dataframes
```

```python
customer_deatil
```

```python
Sales_detail
```

```python
pd.merge(customer_deatil, Sales_detail, left_on='cust_id', right_on='c_id' ) $# by deafault it takes inner join
                                                                            ## if both the primary key and foreign key are simialr we dont have to give any left and right , as by deafult it is inner join
```

```python
pd.merge(customer_deatil,Sales_detail, left_on='cust_id', right_on='c_id', how='left') ## left outer join
```

```python
pd.merge(customer_deatil,Sales_detail, left_on='cust_id', right_on='c_id', how='right') ## right outer join
```

```python
pd.merge(customer_deatil,Sales_detail, left_on='cust_id', right_on='c_id', how='outer') ## full outer join
```

```python
    pd.concat([customer_deatil, Sales_detail], axis=0) ## index join 
```

```python
    pd.concat([customer_deatil, Sales_detail], axis=1) ## index join on row
```
