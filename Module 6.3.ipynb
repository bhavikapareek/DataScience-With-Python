# 4. Compute â€“ Create linear model plot of Length of Membership and Yearly Amount Spent. Does the data fits well in linear plot?

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
x = pd.DataFrame(dataset.iloc[:,3:7].values)
y = pd.DataFrame(dataset.iloc[:,7].values)
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.2, random_state=85)

regressor = LinearRegression()
model = regressor.fit(x_train, y_train)
```

```python
model
```

# Compute â€“ Predict the data and do a scatter plot. Check if actual and predicted data match?

```python
y_pred = regressor.predict(x_test)
plt.scatter(y_test, y_pred, color='pink')
plt.xlabel('y_test')
plt.ylabel('y_pred')
plt.title('Test vs. Pred')
plt.grid(True)
plt.show()
```

# What is the value of Root Mean Squared Error?

```python
from sklearn.metrics import r2_score
r2_score(y_test,y_pred)
```
