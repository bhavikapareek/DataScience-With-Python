# Changing columns name

```python
df=df.rename(columns={'education.num':'education_num', 'marital.status':'marital_status', 'capital.gain':'capital_gain',
                      'capital.loss':'capital_loss', 'hours.per.week':'hours_per_week', 'native.country':'native_country'})
```

```python
df.columns
```

# Selecting columns using dfply

```python
## Selecting specific columns
(df>>
select(X.age, X.workclass, X.occupation, X.education, X.income))
```

# Dropping columns using dfply

```python
## Dropping columns (using not option)
(df>>
select(~X.fnlwgt, ~X.education_num, ~X.capital_gain, ~X.capital_loss))
```

```python
## Dropping columns
(df>>
drop(X.fnlwgt,X.education_num, X.capital_gain, X.capital_loss))
```

# Filter rows- Mask Command


```python
## Filtering records based on conditions
(df>>
mask(X.age>50, X.education=='HS-grad', X.workclass=='Private'))
```

# Sort rows-Arrange Command

```python
## Sorting rows
(df>>
arrange(X.age, X.hours_per_week, ascending=False)>>
head(10))
```

# Adding new columns- Mutate Command


```python
## Creating new column
(df>>
mutate(workclass_education=X.workclass+'-'+X.education, yearly_hours=X.hours_per_week*52))
```

# Group by

```python
## group by data
(df>>
group_by(X.sex))
```

# Summarise Data

```python
## grouping and find the mean age
(df>>
group_by(X.occupation)>>
summarise(mean_age=X.age.mean()))
```
