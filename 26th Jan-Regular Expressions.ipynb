```python
import re #import lib for extracting text
```

```python
nameage='''Janice is 22 and Thron is 33
Gadriel is 44 and Joey is 21'''
```

```python
nameage
```

```python
ages = re.findall(r'\d{1,2}', nameage) # extracting numbers from text
```

```python
ages
```

```python
names=re.findall(r'[A-Z][a-z]*', nameage)  # extracting names from text
```

```python
names
```

```python
agedict={}  ## defining a dictioanry from the both ages and names
x=0
for i in names:
    agedict[i]=ages[x]
    x=x+1
print(agedict)    
```

```python
a='We need to inform him with the latest information'  ## search from text and print the message
if re.search('inform', a):
    print('information is available')
```

```python
a='We need to inform him with the latest information' ## find how many times inform is there in particular text
allinform=re.findall('inform', a)
allinform
```

```python
## find the start and end position of inform
for i in re.finditer('inform', a):
    loc=i.span()
    print(loc)
```

```python
str='sat, mat, rat,hat, pat' ## dont want text starting from h till m and has at
allstr=re.findall('[^h-m]at', str)
for i in allstr:
    print(i)
```

```python
allstr=re.findall('[shmpr]at', str) ## find all letters
for i in allstr:
    print(i)
```

```python
str
```

```python
regex=re.compile('[r]at') ### substitue one value with another
str =regex.sub('food', str)
str
```

```python
## find numbers whose length is 5 and max len is 7
num ='123    1234    12345    123456    1234567   12345678 '
len(re.findall('\d{5,7}',num))
```

```python
## validate phone number
phone= '412-333-2345'
if re.search('\d{3}-\d{3}-\d{4}', phone):
    print('This number is valid')
else:
    print('Check your format')
```

```python
## validate phone number with + sign
phone= '+1-412-333-2345'
if re.search('\+\d{1}-\d{3}-\d{3}-\d{4}', phone):
    print('This number is valid')
else:
    print('Check your format')
```
