# Checking for outliers using Boxplot

```python
sns.boxplot(y='age', data=dataset, color= 'green')
```

```python
sns.boxplot(x='hours.per.week', data=dataset, color= 'red')
```

# Scatterplot

```python
sns.lmplot(data=dataset, x='hours.per.week', y='age', hue='occupation')
```

```python
sns.lmplot(data=dataset, x='hours.per.week', y='age', hue='income', fit_reg=False)
```
