```python
x=input('Enter your Dept:')  ## Dept example- code error(to be checked)
Dept=['Marketing', 'HR', 'Operations', 'Sales', 'Finance']
z=0
for y in range(len(Dept)):
    if Dept[y] == x:
        z=1
        break
    else:
        continue
if z>0:
    print('valid')
else:
    print('not valid')
        
```
